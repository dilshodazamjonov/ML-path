## Model For Predicting the Price range of newly came phone, based on charasteristics of it

## Data Cab be downloaded through this link:
https://www.kaggle.com/datasets/iabhishekofficial/mobile-price-classification/data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


df = pd.read_csv('data/train.csv')
```

```python
df.describe()
```

```python
df.columns
```

```python
input_cols = ['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']
target_col = ['price_range'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[input_cols], df[target_col], test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier 

model1 = RandomForestClassifier(n_estimators=200, max_depth=20)
model1.fit(X_train, y_train)
model1.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model1.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
importances  = model1.feature_importances_
importances
```

```python
res = pd.DataFrame({"feature": input_cols, "importance": importances}).sort_values("importance", ascending=False)
```

```python
res.to_csv('data/outputs/feature_importances.csv')
```

## Testing on the test data.

```python
test_df = pd.read_csv('data/test.csv')
```

```python
valid_cols = ['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc',
       'four_g', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc',
       'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']

input_df = test_df[valid_cols]
```

```python
def predict_and_save(model, model_name):
    res = model.predict(input_df)
    prediction_proba = model.predict_proba(input_df)
    proba_res = prediction_proba.max(axis=1)
    output = pd.DataFrame({
        'id': test_df['id'],
        'prediction': res,
        'probability': proba_res
    })
    try:
        output.to_csv(f'data/outputs/{model_name}_output.csv', index=False)
    except Exception as e:
        raise e
```

```python
predict_and_save(model1, 'rf')
```

## Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier(n_neighbors=9, n_jobs=-1)
model2.fit(X_train, y_train)
model2.score(X_train, y_train), model2.score(X_test, y_test)
```

```python
predict_and_save(model2, 'knn')
```
