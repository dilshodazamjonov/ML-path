## Model For Predicting the Price range of newly came phone, based on charasteristics of it

## Data Cab be downloaded through this link:
https://www.kaggle.com/datasets/iabhishekofficial/mobile-price-classification/data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


df = pd.read_csv('data/train.csv')
```

```python
input_cols = ['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']
target_col = ['price_range'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[input_cols], df[target_col], test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier 

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
importances  = model.feature_importances_
importances
```

```python
pd.DataFrame({"feature": input_cols, "importance": importances}).sort_values("importance", ascending=False)
```
